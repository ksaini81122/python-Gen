```python
import numpy as np
import pandas as pd ##data manipulation and analysis
##data visualization --- charts and plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##reading the dataset
df = pd.read_csv("cust_satisfaction.csv")
df
```

```python
##data cleaning 
##1 -- null values remove 
##2 -- duplicate values remove

##1 null values 
##to check null values 
df.isnull().sum()
##to remove null values
df.dropna(inplace=True) 
# df.isnull().sum()
```

```python
##to check duplicate values
df.duplicated().sum()
##to remove duplicate values
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
##to convert alphabetical values to numerical values
##one hot encoding -- 0 and 1 me 
df["Gender"] = df["Gender"].map({"Male":0,"Female":1})
df["Customer Type"] = df["Customer Type"].map({"Loyal Customer":1,"disloyal Customer":0})
df["Type of Travel"] = df["Type of Travel"].map({"Personal Travel":0,"Business travel":1})
df["Class"] = df["Class"].map({"Business":1,"Eco":2,"Eco Plus":3})
df["satisfaction"] = df["satisfaction"].map({"neutral or dissatisfied":0,"satisfied":1})
```

```python
##pandas indexing  has 2 typees
#1 -- loc -- label based indexing
#2 -- iloc -- position based indexing

# df.loc[0:10,["Gender","Customer Type"]]
df.iloc[0:10,0:3]
```

```python
##handle imbalance data
loyal_customer = df[df["Customer Type"] == 1].loc[0:20000]
disloyal_customer = df[df["Customer Type"] == 0]
```

```python
#axis = 0 -- row
#axis = 1 -- column
balance_df = pd.concat([loyal_customer,disloyal_customer],axis=0)
balance_df
```

```python
##data divide x and y -- 
x = balance_df.drop("Customer Type",axis=1)
y = balance_df[["Customer Type"]]
```

```python
##data ko divide training and testing --80-20 ratio
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,
                                                 random_state=42)
```

```python
##normalise --- data range
from sklearn.preprocessing import StandardScaler,MinMaxScaler
## mean = 0 , std = 1
sd = StandardScaler()
x_train_scaled = sd.fit_transform(x_train)
x_test_scaled = sd.transform(x_test)
```

```python
##ANN architure 
##il , hl , ol 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
##ann model define 
model = Sequential()
##input layer ---32 , 64 , 128 
model.add(Dense(68,activation="relu",input_dim = x_train_scaled.shape[1]))
##hidden layer 
model.add(Dense(32,activation="relu"))
model.add(Dense(24,activation="relu"))
model.add(Dense(12,activation="relu"))
##output layer 
model.add(Dense(1,activation="sigmoid"))

#model compile 
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
model.summary()
```

```python
##formula of calculate weights
##tp = (no of features * no of neurons ) + no of neurons
TP = 68 * 32 + 32
TP
```

```python
##train the model 
history = model.fit(x_train_scaled,y_train,epochs=10,
                    validation_data=(x_test_scaled,y_test))
history
```
